```python
rfile = open("/Users/reejungkim/Documents/geo_countries.json", 'r', encoding='utf-8').read()
jsonData = json.loads(rfile)
folium.GeoJson(jsonData, name='json_data').add_to(geo_map)

geo_map
```

```python
choropleth = folium.Choropleth(
    geo_data= jsonData,
    name='choropleth',
    data=df,
    columns=['Country', 'Quantity'],
    key_on='properties.admin',
    fill_color= 'YlGn',
    fill_opacity=0.7,
    line_opacity=0.2,
    legend_name='Quantity ordered',
    highlight=True,
    line_color='blue'
).add_to(geo_map)

folium.LayerControl(collapsed=True).add_to(geo_map)
geo_map
```

```python
import requests

url = 'https://raw.githubusercontent.com/python-visualization/folium/master/examples/data'
us_states = f'{url}/us-states.json'

geo_json_data = json.loads(requests.get(us_states).text)
geo_json_data
```

```python
m = folium.Map([43, -100], zoom_start=4)
folium.GeoJson(geo_json_data).add_to(m)

m
```

```python
m = folium.Map([43, -100], zoom_start=4)

folium.GeoJson(
    geo_json_data,
    style_function=lambda feature: {
        'fillColor': 'green' if 'e' in feature['properties']['name'].lower() else '#ffff00',
        'color': 'black',
        'weight': 2,
        'dashArray': '5, 5'
    }
).add_to(m)

m
```
